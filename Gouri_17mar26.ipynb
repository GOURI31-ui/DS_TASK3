```python
# Created A list named colours
#1.Add " yellow" tothe end of the list

colours=["red","green","blue"]
colours.append("yellow")
print(colours)
```

```python
#2. Insert purple at the second of the list
colours=["red","green","blue"]
colours.insert(1,"purple")
print(colours)
```

```python
#3. reverse the entire list 

colours.insert(0,"yellow")
colours.insert(1,"blue")
colours.insert(2,"green")
colours.insert(3,"red")

print(colours)
```

```python
#4.print the final list and its total length
colours.reverse()
print(colours)
print("Final list :",colours)
print("length of list:",len(colours))
```
